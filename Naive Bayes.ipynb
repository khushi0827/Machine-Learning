```python
# importing librarirs
import pandas as pd # pandas (pd): Used for handling structured data (tables, CSVs, Excel files, etc.).
import numpy as np # numpy (np): Provides numerical operations, arrays, and mathematical functions.
from sklearn.preprocessing import LabelEncoder #Preprocessing Tools = LabelEncoder: Converts categorical labels into numeric values.
from sklearn.naive_bayes import CategoricalNB , GaussianNB , BernoulliNB , MultinomialNB
# Categorical features (like colors, categories),Continuous features (like height, weight),Binary features (0/1, True/False),Count-based features (like word frequencies in text)
from sklearn.model_selection import train_test_split
# train_test_split: Divides your dataset into training set (used to train the model) and test set (used to evaluate performance).
```

```python
go=pd.read_csv(r"C:\Users\user\Downloads\golf.csv")  #  loading your dataset into Python
```

```python
go
```

```python
LE = LabelEncoder() # Create a LabelEncoder object
for i in go.columns: # Loop through each column in the DataFrame
    go[i]=LE.fit_transform(go[i]) # Apply Label Encoding to each column
```

```python
x=go.drop('Play',axis=1)
y=go.Play  #  separating your features (X) and target (y) from the dataset so you can train a machine learning model
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.4)  # splitting your dataset into training and testing 
```

```python
CNB= CategoricalNB() # creates an instance of the Categorical Naive Bayes classifier
```

```python
CNB.fit(xtrain,ytrain)  # fit method is used to train the model
```

```python
CNB.score(xtest,ytest) # calculates the accuracy of the model.Accuracy = (Correct predictions ÷ Total predictions)
```

```python
CNB.score(xtrain,ytrain) # accuracy of the model on the training set.
```
