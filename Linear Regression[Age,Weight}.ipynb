```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder , StandardScaler , MinMaxScaler 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Starting with Linear Regression [Basic]

```python
Age = np.concatenate([np.arange(18,60,2) , np.array([64])],axis=0)
Weight  = np.concatenate([np.arange(50,92,2) , np.array([50])],axis=0)
```

```python
print(Age)
print(Weight)
```

```python
plt.scatter(Age,Weight)
```

```python
x=Age.reshape(22,1)
y=Weight
```

```python
xtrain , xtest , ytrain , ytest  = train_test_split( x , y , test_size=0.3)
```

```python
print('X Data For Training : ' , len(xtrain))
print('Y Data For Training : ' , len(ytrain))
print('X Data For Testing : ' , len(xtest))
print('Y Data For Testing : ' , len(ytest))
```

```python
LR = LinearRegression()
```

```python
LR
```

```python
LR.fit(xtrain,ytrain)
```

```python
print('Actual',ytrain)
print('predict',LR.predict(xtrain))
```

```python
LR.score(xtrain,ytrain)
```

```python
print('X daya', xtest)
print('Prediction',LR.predict(xtest))
```

```python
print(Age)
print(Weight)
```

```python
LR.score(xtest,ytest)
```

```python
M=LR.coef_
```
