# SVM Algorithm 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.pairplot(iris,hue='species',palette='viridis') # pairplot bnakr color k hisab se alag kiya h 
```

```python
iris.columns
```

```python
x=iris[['petal_length','petal_width']]
y=iris.species
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
SVM=SVC(kernel='linear')
```

```python
SVM.fit(xtrain,ytrain)
```

```python
SVM.score(xtrain,ytrain)
```

```python
SVM.score(xtest,ytest)
```

```python
from sklearn.datasets import make_blobs ,make_circles
```

```python
x,y=make_blobs(n_samples=200,centers=4,random_state=42)
```

```python
y=np.logical_xor(y==0,y==1)
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y,palette='viridis')
```

```python
SVM=SVC(kernel='poly')
```

```python
SVM.fit(xtrain,ytrain)
```

```python
SVM.score(xtrain,ytrain)
```

```python
SVM.score(xtest,ytest)
```

```python
x,y=make_circles(n_samples=600,noise=0.15,factor=0.4,random_state=42)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y,palette='viridis')
```

```python
SVM=SVC(kernel='rbf')
```

```python
SVM.fit(xtrain,ytrain)
```

```python
SVM.score(xtrain,ytrain)
```

```python
SVM.score(xtest,ytest)
```
